Для разбора возьмем приложение ozon.

1. Уведомления.

Гипотеза : Всплывающее уведомление о скидках/акциях о товарах в интересующих пользователя категориях увеличит повторные возвращения пользователя и количество совершаемых им покупок.

Тестируем : На всех пользователях ранее установивших наше приложение.

Метрики: Количество повторных посещений приложения и количество пользователей совершивших покупки в предложенных категориях.

2. Remote config.

Гипотеза : Замена рекламного банера на стороние ресурсы  на витрину с актуальнными предложениями и акциями увеличит комфорт/информированность пользователей и будет мотивировать их совершать покупки внутри приложения.

Тестируем : Всех пользователей.

Метрики : Количество переходов с приветственнного экрана на страницу конкрентного предложения. Количество временни проводимого в приложении на изучение всех рекомендаций. 
Сравниваем с активностью пользоваталей перешедших по рекламному баннеру стороних ресурсов.

3. Уведомления внутри приложения -in-app.

Гипотеза : Для пользователей добавляющих товар в корзину всплывающее уведомление будет предлагать изучить рекомендованный сопутствующий товар.

Тестируем : На пользователях дoбавляющих товар в корзину.

Метрика : Количество пользователей которые изучат или купят сопутствующий товар.
