### Задание 2

1. Изменение дизайна приветственного экрана. Если посетители сайта пользуются увеличением масштаба отображенния, то картинка получается рванной и не структурированнной. Большая часть экрана занята баннером афиши с неконкретными предложениями. Это созадет ощуение посредственности ресурса
и снижает доверие посетителя к платформе/желанние что то покупать.
Можно изменить выводимые элементы, избавится от лишнней неактульной информации для посетителя - перенеся категории товаров на левый борд с всплывающщим окном. Тогда это позволит сконцентрировать внимание на рекламных и акционннных предложениях, что даст посетителю ощущение полезности актальных предложений и экономии его времени на ознакомление с ненужнным спамом.

2. Нужно улучшить рекомендации посетителю, как общего характера для новых пользователей, так и авторизированному с личными предпочтениями. Нужно предлагать действительнно полезнные и выгодные по цене товары, пусть из категории дешевых товаров или постоянно нужных расходников. Сегодня у любого пользователя нет проблем с доступом к информации и каждый может уточнить цены на рекомендованные предложенния, сравннив их с ценами конкурентов. Потому использовать данную систему нужно в целях завоевания доверия пользователя к алгоритмам платформы, что будет повышать лояльность клиента. Понимая что алгоритмы рабоют на пользу покупателю, он будет это ценить и менньше временни будет тратить на проверку информации на ресурсах конкурентов.

3. Нужнно изменить систему фильтрации при подборе товара. Она слишком вытянута и заставляет клиента скролить экран, отвлекаясь от первичных предложений. Так же это тратит время, что воспринимается негативно. Нужно исправить технические проблемы отображения при выборе нескольких параметров фильтрации.

4. Стоит изменить амбассадора компании. В условиях событий последних лет выбор Галустяна как амбасадора компании выглядит неуместно, создает негативно несерьезный образ для однного из крупнейших дискаунтеров страны. Можно проработать более актуальные для сегодняшнего дя персоны или попробовать рокировку на Светлакова.
