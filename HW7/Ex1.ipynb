Проблема.
После анализа количественных данных о продукте выяснилось что  привлеченные в рамках конкретных маркетинговых программ клиенты проявляют низкую лояльность к продукту. После покупки конкретного продукта у такие пользователи редко повторно посещают платформу, а при посещении проводят мало времени в нашем продукте. 

### Гипотеза 1.

1. Для повышения лояльности пользователей внедрить систему личных пуш уведомлений с рекомендациями для конкретных пользователей на основа их поисковых запросов.
2. Основная метрика - количество переходов по ссылке на странницу рекомендованого в пуш уведомлении продукта.
3. Вторичная метрика - количество времени проводимого пользовтаелем на нашей платформе.
4. Ожидаемый эффект - для пользователей подключенных к системе пуш уведомлений ожидается рост повторных посещений платформы на 30%.
5. Целевая группа - пользователи приходящие на платформу в рамках конкретных маркетинговых компаний с выгодными условиями относительно рынка. 
6. Размер выборки для теста - 50% от целевой группы в тестовой и 50% в контрольной группе.
7. Время теста. В зависимости от ежедневного траффика в рамках выделенных групп запускаем тест длительностью 4 недели.

### Гипотеза 2.

1. Для повышения лояльности пользователей проведем маркетинговую программу мзучения продуктовых предпостений пользовтаеля. Будет предложено пройти анкетирование с продуктовыми предпочтениями пользователя для настройки индивидуальной системы рекомендаций. В качестве вознаграждения пользователю будут начислены бонусные баллы с огранниченным периодом действия для второй и последующих покупок.
2. Основная метрика - количество повторных покупок.
3. Вторичная метрика - количество повторных посещений платформы для изучения личных рекомендаций, переход на карточку рекомендованного продукта. 
4. Ожидаемый эффект - лояльность пользователей к платформе и количество повторных покупок вырастет на 20%. 
5. Целевая группа - пользователи совершившие первую покупку на платформе.
6. Размер выборки для теста - 25% от целевой группы в тестовой и 75% в контрольной группе.
7. Время теста - В зависимости от условий и времени активации бонусных баллов - тест будет длиться 8 недель. 

Определяем критические значения для основных метрик, если эксперимент окажет негативное влияние на эти метрики и превысит допустимый диапазон,  то мы остановим эксперимент. Для избежания проблемм подглядывания мы не будем делать никаких выводов относиельно похитивных изменений метрик до полного окончания эксперименнта и оценки всех факторов влияюих на итоговый результат. О проведении эксперимента будут уведомлены все продуктовые команды. 
Если тест дал положительный эффект, масштабируем на всех пользователей.
