#### Задание 2
1. Гипотеза изменения приветственнного экрана. Если изменить акцент внимания посетителей на лучшие предложения и убрать отвлекающие внимание ссылки - это увеличит количество переходов на страницы маркетинговых акций и конкретных товаров на 20%. Тестируем на всех пользователях.
Метрика - количество переходов со стартовой страницы на страницу конкретного товара.

2. Гипотеза оптимизации  рекомендаций. Внедрение системы персональных рекомендаций в зависимости от предпочтений просмотра определенных групп товаров для авторизованых пользователей. Тестируем авторизованных пользователей совершивших хотя бы одну покупку и участвующих в программах лояльности.
Метрика - Частота повторного посещения приложения клиентом.

3. Изменение сиситемы фильтрации товаров. Улучшение системы подбора товаров увеличит среднее время нахождения потребителя на сайте и количество просмотренных карточек товара. 
Тестируем на всех категориях посетителей.
Метрика - количество переходов на страницу конкретного товара из списка отсортированных товаров.
