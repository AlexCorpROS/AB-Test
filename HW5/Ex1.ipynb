#### Задача 1

Так как у нас выбрана метрика конверсии в покупку(вид метрики доли), то критерием для проверки гипотез выбираем Z-критерий долей, либо Хи-квадрат на однородность.

Z-критерий для разности долей. Формула Z-критерия:
Z = (p1 - p2) / sqrt(p * (1 - p) * (1/N1 + 1/N2))

где p = (n1 * p1 + n2 * p2) / (n1 + n2)

p1 и p2 - конверсии в группах, N1 и N2 - количество юзеров в группах, которые заходили на сайт в период эксперимента, n1 и n2 - количество юзеров, которые совершили хотя бы одну покупку за период эксперимента.

```python
# Рассчитаем z-критерий для разности долей

p1, p2, n = 164, 228, 15550

p =  (p1 * p1/n + p2 * p2/n) / (p1 + p2)

Z = (p1/n - p2/n) / ((p * (1 - p) * (1/n + 1/n))**0.5)
Z
```

Для α=5% в двухсторонней альтернативе критическое значение Z-критерия  равно 1.96. Расчетное значение Z меньше  1.96 ,  мы отвергаем нулевую гипотезу. Внесенные изменения дали статистически значимый эффект на уровень конверсии в покупку. Рекомендовано использовать новую форму заказа.

```python
# Рассчитаем z-критерий используя библиотеку statsmodels

import numpy as np
import statsmodels.stats.proportion as smp

alpha = 0.05

purchases = np.array([164, 228])
users = np.array([15550, 15550])

conversion = purchases/users
zstat, pvalue = smp.proportions_ztest(count=purchases, nobs=users, value=0, alternative='two-sided')
zstat, pvalue
```

P-value меньше заданного значения альфа в 0.05 - разница между группами стат значима. 

```python
# Посчитаем доверительный интервал для z-критерия пропорций. p2-p1 +-(p1*(1-p1)/n1 + p2*(1-p2)/n2))

p1, p2, n = 164, 228, 15550


P1 = p1/n
P2 = p2/n
delta = p1/n - p2/n
deviations = 1.96*(((P1*(1-P1))/n + (P2*(1-P2))/n)**0.5)
interval = [(delta - deviations)*100, (delta + deviations)*100]
interval

# 0 не входит в интервал, значит между группам есть статистически значимые различия
```
